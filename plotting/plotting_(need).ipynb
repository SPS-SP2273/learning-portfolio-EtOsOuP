# 1 A simple plot

```python
from matplotlib import pyplot as plt
```

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]          # X-axis values
y_1 = [0, 2, 4, 6, 8, 10, 12]      # Y-axis values 
y_2 = [0, 3, 6, 9, 12, 15, 18]     # Second set of Y values

# Let's start plotting
plt.plot(x, y_2, label='Y values', # Command to start plotting. can have multiple y data
         c='black', ls='dashed', lw='2') # X-axis data, Y-axis data, legend, line colour,

plt.plot(x, y_1, label='Y values',
         color= 'red', linestyle='dashed') #can differentiate btw sets of data liddis


plt.xlabel('x-values')             #axis labels for X
plt.ylabel('y-values')
plt.title('X vs Y')                #plt= plot? so plot title is this this
plt.grid(alpha=.50)                #grid line thickness? from .1 to .9, .9 thickest
plt.legend(loc='upper left')       #location of legend
```

```python
plt.show() #good to include this in line so code will work even in a non-responsive environment
```

## 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75 #error bars need to be defined first

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',              #each line needs to have its own plt.plot
         color='blue')
plt.legend()   #is blank cuz the label values are alr defined.                              
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2. need to hv


```

# 3 Saving to disc

```python
x = np.linspace(0, 10, 100) #x value start, stop, 'smooth value of the curve'
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

function: savefig() allows saves to disk
We specify the format with the extension (e.g., filename.pdf) of the file name,
We specify resolution by use dots-per-inch (dpi).

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-5, 5]) #y- axis lower and upper limit
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') 
#horizontal line with value zero, start x value, end x value, 
```

## 4.2 Styles

matplot lib has standard style templetes that can make plot more asthetic

```python
plt.style.use('fivethirtyeight')
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-5, 5]) #y- axis lower and upper limit
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') 
#horizontal line with value zero, start x value, end x value, 
```

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24) #skiprows = skip rows number from 1 to 24
#do this if there are header rows or irrelevant info at the beginning of the file

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
