```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow', #'o'rep a circle marker in matplotlib
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#range increment is 1 so using this to plot = jaggared list and append to generate the 1k points (prac)
x = np.linspace(0, 50, 1000)  # Generate 1000 points between 0 and 50
y = np.exp(-x / 10) * np.sin(x)  # Calculate y values for the given function

plt.plot(x, y, label=r'$y = e^{-\frac{x}{10}} \sin(x)$')
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange', rotation=0)
plt.ylim([-1, 1])
plt.title('$y = e^{-x/10} \sin(x)$', fontsize=20)
plt.grid(alpha=.25)
plt.legend().set_visible(True)
plt.show()
```
